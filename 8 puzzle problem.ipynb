## In this code I make a puzzle game. Possible moves depend on the position of the space. Based on the heuristic value only the option which has least misplaced tiles is selected from the given set of option. The code is keep on running until goal state is not found.
### Goal state is given below
#### 0         1        2
#### 3         4        5
#### 6         7        space 


```python
def space_pos(gameList):
    for i,value in enumerate(gameList):
        if value == ' ':
            return i

def isMatch(test):
    for i, item in enumerate(test):
        if i==8 and item==' ': #last test case
            return True
        if i != item:
            return False
    return True

def printPuzzle(temp):
    for i, value in enumerate(temp):
        if i%3 == 0:
            print()
        print(value, end='    ')

def misplacedTiles(temp):
    count = 0
    for i, value in enumerate(temp):
        if i == 8 and value != ' ':
            count += 1
        elif i != value:
            count += 1
    return count

def solve_puzzle(queue):
    record = list(queue)
    temp = queue.pop(0)
    while(not isMatch(temp)):
        sp_pos = space_pos(temp)
        
        printPuzzle(temp)
        print()
        print()

        if sp_pos == 4:
            #space goes up
            t = list(temp)
            t[4] = t[1]
            t[1] = ' '
            
            flag = False
            if t not in record:
                queue.append(t)
                record.append(t)
                flag = True
            
                misTiles = misplacedTiles(t)
            
            #space goes right
            t = list(temp)
            t[4] = t[5]
            t[5] = ' '
            
            if t not in record:
                if flag:
                    misTiles2 = misplacedTiles(t)
                    if misTiles > misTiles2:
                        queue.pop()
                        queue.append(t)
                        record.pop()
                        record.append(t)
                        misTiles = misTiles2
                else:
                    queue.append(t)
                    record.append(t)
                    flag = True
                    
                    misTiles = misplacedTiles(t)
                
            #space goes down
            t = list(temp)
            t[4] = t[7]
            t[7] = ' '
            
            if t not in record:
                if flag:
                    misTiles2 = misplacedTiles(t)
                    if misTiles > misTiles2:
                        queue.pop()
                        queue.append(t)
                        record.pop()
                        record.append(t)
                        misTiles = misTiles2
                else:
                    queue.append(t)
                    record.append(t)
                    flag = True
                    
                    misTiles = misplacedTiles(t)
            

            #space goes left
            t = list(temp)
            t[4] = t[3]
            t[3] = ' '
            
            if t not in record:
                if flag:
                    misTiles2 = misplacedTiles(t)
                    if misTiles > misTiles2:
                        queue.pop()
                        queue.append(t)
                        record.pop()
                        record.append(t)
                        misTiles = misTiles2
                else:
                    queue.append(t)
                    record.append(t)
                    flag = True
                    
                    misTiles = misplacedTiles(t)

        elif sp_pos%2 == 0:  #two options available for space
            swapPos1 = 1
            swapPos2 = 3 
            if sp_pos == 2:
                swapPos1 = -1
            elif sp_pos == 6:
                swapPos2 = -3
            elif sp_pos == 8:
                swapPos1 = -1
                swapPos2 = -3
            
            #option one
            t = list(temp)
            t[sp_pos] = t[sp_pos + swapPos1]
            t[sp_pos + swapPos1] = ' '
            
            flag = False
            if t not in record:
                queue.append(t)
                record.append(t)
                flag = True
            
                misTiles = misplacedTiles(t)

            #option two
            t = list(temp)
            t[sp_pos] = t[sp_pos + swapPos2]
            t[sp_pos + swapPos2] = ' '
            
            if t not in record:
                if flag:
                    misTiles2 = misplacedTiles(t)
                    if misTiles > misTiles2:
                        queue.pop()
                        queue.append(t)
                        record.pop()
                        record.append(t)
                        misTiles = misTiles2
                else:
                    queue.append(t)
                    record.append(t)
                    flag = True
                    
                    misTiles = misplacedTiles(t)

        else: #three options available for space
            swapPos1 = -1
            swapPos2 = 1 
            swapPos3 = 3
            if sp_pos == 3:
                swapPos1 = -3
            elif sp_pos == 5:
                swapPos1 = -3
                swapPos2 = -1
            elif sp_pos == 7:
                swapPos3 = -3

            #option one
            t = list(temp)
            t[sp_pos] = t[sp_pos + swapPos1]
            t[sp_pos + swapPos1] = ' '
            
            flag = False
            if t not in record:
                queue.append(t)
                record.append(t)
                flag = True
            
                misTiles = misplacedTiles(t)
            
            #option two
            t = list(temp)
            t[sp_pos] = t[sp_pos + swapPos2]
            t[sp_pos + swapPos2] = ' '
            
            if t not in record:
                if flag:
                    misTiles2 = misplacedTiles(t)
                    if misTiles > misTiles2:
                        queue.pop()
                        queue.append(t)
                        record.pop()
                        record.append(t)
                        misTiles = misTiles2
                else:
                    queue.append(t)
                    record.append(t)
                    flag = True
    
                    misTiles = misplacedTiles(t)

            #option three
            t = list(temp)
            t[sp_pos] = t[sp_pos + swapPos3]
            t[sp_pos + swapPos3] = ' '
            
            if t not in record:
                if flag:
                    misTiles2 = misplacedTiles(t)
                    if misTiles > misTiles2:
                        queue.pop()
                        queue.append(t)
                        record.pop()
                        record.append(t)
                        misTiles = misTiles2
                else:
                    queue.append(t)
                    record.append(t)
                    flag = True
                    
                    misTiles = misplacedTiles(t)
                
        
        temp = queue.pop(0)
        
    printPuzzle(temp)
        

gameInput = input('Enter starting values for the game (format: Separated by commas): ')
gameInput = gameInput.split(',')

queue = []
gameInitial = []
print(gameInput)
for i in gameInput:
    if i == ' ':
        gameInitial.append(i)
    else:
        gameInitial.append(int(i))
    
queue.append(gameInitial)
solve_puzzle(queue)
```

#### Enter initial state of the game.
#### Each value must be separated by space.
#### Example Input: 0,1,2, ,3,5,6,4,7
#### Final State: 0,1,2,3,4,5,6,7,
